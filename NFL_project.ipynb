```python
#Import libraries

import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import numpy as np

#Read the data 

attendance_df = pd.read_csv('/Users/adrianaquinones/Downloads/archive/attendance.csv')
games_df = pd.read_csv('/Users/adrianaquinones/Downloads/archive/games.csv')
standings_df = pd.read_csv('/Users/adrianaquinones/Downloads/archive/standings.csv')
```

```python
#Missing Data fo attendance_df 
for col in attendance_df:
    missing_values = np.mean(attendance_df[col].isnull())
    print('{} - {}%'.format(col, missing_values))
```

```python
#Missing Data for game_df 
for col in games_df:
    games_values = np.mean(games_df[col].isnull())
    print('{} - {}%'.format(col, games_values))
```

```python
#Missing Data fo standings_df 
for col in standings_df:
    standings_values = np.mean(standings_df[col].isnull())
    print('{} - {}%'.format(col, standings_values))
```

```python
attendance_df['weekly_attendance'] = attendance_df['weekly_attendance'].fillna(0).astype('int64')
print(attendance_df)
```

```python
attendance_df.rename(columns={'team': 'team_state'}, inplace=True)
print(attendance_df)
```

```python
standings_df.rename(columns={'team_state': 'team_city'}, inplace=True)
print(standings_df.head())
```

```python
#Calculate total attendance for each team by year 
attendance_total = attendance_df.groupby(['team_name', 'year'])['total'].sum().reset_index()
# Plot total attendance for each team across all years
plt.figure(figsize=(14, 8))
sns.barplot(data= attendance_df, x ='team_name', y ='total', hue = 'team_name', palette = 'rocket')
plt.xlabel('Team')
plt.ylabel('Total Attendance (in millions)')
plt.title('Total Attendance by Team Across All Years')
plt.xticks(rotation=90)
plt.show()
```

```python
#Calculate home and away attendance for each team by year
home_away = attendance_df.groupby(['team_name', 'year'])[['home', 'away']].sum().reset_index()
# Plot home vs away attendance for each team
home_away_melted = home_away.melt(id_vars=['team_name', 'year'], value_vars=['home', 'away'], var_name='Type', value_name='Attendance')

plt.figure(figsize=(14, 8))
sns.barplot(data = home_away_melted, x='team_name', y='Attendance', hue='Type', palette='coolwarm')
plt.xlabel('Team')
plt.ylabel('Attendance')
plt.title('Home vs Away Attendance for Each Team')
plt.xticks(rotation=90)
plt.legend(title='Type')
plt.show()
```

```python
# Merge standings and attendance data
# Calculate average performance and home attendance for each team by year
merged_data = pd.merge(standings_df, attendance_df, on=['team_name', 'year'])
performance_vs_attendance = merged_data.groupby(['team_name', 'year'])[['wins', 'home']].mean().reset_index()
# Performance and Attendance
plt.figure(figsize=(14, 8))
for team in performance_vs_attendance['team_name'].unique():
    team_data = performance_vs_attendance[performance_vs_attendance['team_name'] == team]
    plt.plot(team_data['year'], team_data['home'], marker='o', label=team)
    
plt.xlabel('Year')
plt.ylabel('Home Attendance (Millions)')
plt.title('Trends in Home Attendance Over Time for Each Team')
plt.legend(title='Team', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.show()
```
